

# Business Questions and Hypotheses for Superstore 2023 Dataset Analysis

## Univariate Analysis

### Sales and Discounts
1. **What is the distribution of sales over time?**
   - Are there high and low sales seasons?

2. **What is the distribution of applied discounts?**
   - What percentage of products are sold at a discount?

### Profit and Profitability
3. **What is the distribution of profit per sale?**
   - What are the median and interquartile range of profit?

4. **What percentage of orders are profitable vs. non-profitable?**

### Segment and Region
5. **What is the proportion of sales by segment (Consumer, Corporate, Home Office)?**
   - Does any segment dominate sales?

6. **What is the distribution of sales by region?**
   - Is there a region that represents the majority of sales?

### Product Category and Sub-Category
7. **What is the distribution of sales by product category (Furniture, Office Supplies, Technology)?**

8. **What is the distribution of sales by product sub-category?**
   - Are there sub-categories that sell significantly more than others?

### Shipping Times
9. **What is the distribution of actual shipping time vs. scheduled shipping time?**
   - How many shipments are on time, early, or late?

## Bivariate Analysis

### Relationship Between Sales and Discounts
10. **Is there a correlation between applied discount and sales?**
    - Hypothesis: Higher discounts lead to higher sales.

### Profitability and Discounts
11. **How do applied discounts affect product profitability?**
    - Hypothesis: Higher discounts can reduce profitability.

### Segment and Profitability
12. **What is the profitability by segment?**
    - Hypothesis: Some segments may be more profitable than others.

### Region and Profitability
13. **How does profitability vary by region?**
    - Hypothesis: Profitability may be higher in some regions due to local economic factors.

### Product Category and Profitability
14. **What is the average profitability by product category?**
    - Hypothesis: Technology may be more profitable than other categories.

### Product Sub-Category and Discounts
15. **Which product sub-category receives the most discounts?**
    - Hypothesis: Some sub-categories, such as furniture, may have higher discounts.

### Shipping Method and Customer Satisfaction
16. **How does the shipping method influence customer satisfaction?**
    - Hypothesis: Faster shipping results in higher customer satisfaction.

### Shipping Status and Shipping Time
17. **What is the relationship between shipping status (on time, late) and actual shipping time?**
    - Hypothesis: Late shipments tend to have a longer actual shipping time than scheduled.

### Quantity of Products and Total Sales
18. **Is there a relationship between the quantity of products ordered and total sales?**
    - Hypothesis: Orders with more products tend to have higher total sales.

### Sales Forecast Accuracy
19. **How accurate are the sales forecasts compared to actual sales?**
    - Hypothesis: Sales forecast accuracy may vary by season and product category.

## Specific Questions for Tableau
20. **How do sales and profitability vary throughout the year by region and segment?**

21. **What are the best-selling and most profitable products in each category?**

22. **Which cities generate the most sales, and what is their average profitability?**

23. **Are there sales patterns on specific days of the week or months of the year?**
